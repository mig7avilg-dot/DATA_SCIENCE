```python
# Determine the number of components to explain 95% variance
n_components_95 = (explained_variance_ratio_cumulative < 0.95).sum() + 1
print(f"Number of components to explain 95% variance: {n_components_95}")

# Apply PCA with the selected number of components
pca_optimal = PCA(n_components=n_components_95)
digits_pca_optimal = pca_optimal.fit_transform(digits.data)

print(f"Shape of PCA transformed data with {n_components_95} components:", digits_pca_optimal.shape)
```

```python
# Calculate explained variance for a range of components
pca_full = PCA()
pca_full.fit(digits.data)

explained_variance_ratio_cumulative = pca_full.explained_variance_ratio_.cumsum()

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio_cumulative) + 1), explained_variance_ratio_cumulative, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()
```
