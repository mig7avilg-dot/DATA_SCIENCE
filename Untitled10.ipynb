```python
from google.colab import drive
drive.mount('/content/drive')
import tifffile as tf
import numpy as np

ruta_carpeta = '/content/drive/MyDrive/PROYECTO1/'

# Nombres exactos de tus archivos
archivo_control = 'Control.tif'
archivo_mutatnt1 = 'Mutatnt1.tif'
archivo_mutatnt2 = 'Mutatnt2.tif'

# Combinar la ruta con el nombre de cada archivo para obtener la ruta completa
ruta_completa_control = ruta_carpeta + archivo_control
ruta_completa_mutatnt1 = ruta_carpeta + archivo_mutatnt1
ruta_completa_mutatnt2 = ruta_carpeta + archivo_mutatnt2

# Cargar cada archivo usando tf.imread()
imagen_control = tf.imread(ruta_completa_control)
imagen_mutatnt1 = tf.imread(ruta_completa_mutatnt1)
imagen_mutatnt2 = tf.imread(ruta_completa_mutatnt2)

# Imprimir la forma de las matrices para verificar la carga
print(f'Archivo "Control.tif" cargado. Forma: {imagen_control.shape}')
print(f'Archivo "Mutatnt1.tif" cargado. Forma: {imagen_mutatnt1.shape}')
print(f'Archivo "Mutatnt2.tif" cargado. Forma: {imagen_mutatnt2.shape}')
```

```python
import numpy as np

def min_max_normalize(image_stack):
    """Normaliza un conjunto de imágenes (stack) a un rango [0, 1]."""
    # Se convierte el stack a tipo float para evitar problemas de casting
    image_stack = image_stack.astype(np.float64)

    # Se calculan los valores mínimo y máximo de todo el stack de imágenes
    min_val = image_stack.min()
    max_val = image_stack.max()

    # Se aplica la fórmula de normalización
    normalized_image = (image_stack - min_val) / (max_val - min_val)
    return normalized_image

# Aplica la función a cada imagen cargada
imagen_control_norm_minmax = min_max_normalize(imagen_control)
imagen_mutatnt1_norm_minmax = min_max_normalize(imagen_mutatnt1)
imagen_mutatnt2_norm_minmax = min_max_normalize(imagen_mutatnt2)
```

```python
from sklearn.preprocessing import StandardScaler

def standardize_image(image_stack):
    """Estandariza un conjunto de imágenes (stack) usando StandardScaler."""
    # Guarda la forma original de la imagen (frames, alto, ancho)
    original_shape = image_stack.shape

    # Redimensiona el stack a 2D (pixels, 1) para que StandardScaler lo pueda procesar
    reshaped_image = image_stack.reshape(-1, 1)

    # Crea y entrena el escalador
    scaler = StandardScaler()
    standardized_image = scaler.fit_transform(reshaped_image)

    # Redimensiona la imagen estandarizada a su forma original
    return standardized_image.reshape(original_shape)

# Aplica la función a cada imagen cargada
imagen_control_stand = standardize_image(imagen_control)
imagen_mutatnt1_stand = standardize_image(imagen_mutatnt1)
imagen_mutatnt2_stand = standardize_image(imagen_mutatnt2)
```

```python
import matplotlib.pyplot as plt

def plot_frames(image_stack, title, num_frames=3):
    """
    Grafica una serie de fotogramas representativos de un stack de imágenes.

    Args:
        image_stack (np.array): El stack de imágenes (3D).
        title (str): El título para la figura.
        num_frames (int): El número de fotogramas a graficar.
    """
    # Determina qué fotogramas graficar (e.g., inicio, medio, fin)
    total_frames = image_stack.shape[0]
    frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)

    fig, axes = plt.subplots(1, num_frames, figsize=(15, 5))
    fig.suptitle(title, fontsize=16)

    for i, idx in enumerate(frame_indices):
        frame = image_stack[idx]
        ax = axes[i]
        im = ax.imshow(frame, cmap='gray')
        ax.set_title(f'Fotograma {idx}')
        ax.axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Ejemplo de uso:
# Graficar 3 fotogramas de la imagen de control normalizada con Min-Max
plot_frames(imagen_control_norm_minmax, 'Control - Normalización Min-Max')

# Graficar 3 fotogramas de la imagen de control estandarizada
plot_frames(imagen_control_stand, 'Control - Estandarización')

# Graficar 3 fotogramas de la imagen mutantnt1 normalizada con Min-Max
plot_frames(imagen_mutatnt1_norm_minmax, 'Mutatnt1 - Normalización Min-Max')

# Graficar 3 fotogramas de la imagen mutatnt1 estandarizada
plot_frames(imagen_mutatnt1_stand, 'Mutatnt1 - Estandarización')

# Graficar 3 fotogramas de la imagen mutantnt2 normalizada con Min-Max
plot_frames(imagen_mutatnt2_norm_minmax, 'Mutatnt2 - Normalización Min-Max')

# Graficar 3 fotogramas de la imagen mutatnt2 estandarizada
plot_frames(imagen_mutatnt2_stand, 'Mutatnt2 - Estandarización')
```

2. Preprocess the Data
Flatten each frame and construct a matrix X with shape (n_frames, n_pixels) for each normalization method. Then create a label vector:

label = 0 → control embryo
label = 1 → mutant 1
label = 2 → mutant 2
This step will prepare the input data for dimensionality reduction.

🧪 Experiment: Try running all dimensionality reduction methods with both versions of the input (X_minmax and X_standard) and compare how they affect embeddings and separability.

```python
# Get the shape of a single frame (height * width)
n_pixels = imagen_control_norm_minmax.shape[1] * imagen_control_norm_minmax.shape[2]

# Concatenate all Min-Max normalized stacks and flatten each frame
# .reshape(-1, n_pixels) flattens each frame and stacks them vertically
X_minmax = np.concatenate(
    (imagen_control_norm_minmax.reshape(-1, n_pixels),
     imagen_mutatnt1_norm_minmax.reshape(-1, n_pixels),
     imagen_mutatnt2_norm_minmax.reshape(-1, n_pixels)),
    axis=0
)

# Concatenate all standardized stacks and flatten each frame
X_standard = np.concatenate(
    (imagen_control_stand.reshape(-1, n_pixels),
     imagen_mutatnt1_stand.reshape(-1, n_pixels),
     imagen_mutatnt2_stand.reshape(-1, n_pixels)),
    axis=0
)

print(f"Shape of X_minmax: {X_minmax.shape}")
print(f"Shape of X_standard: {X_standard.shape}")
```

```python
# Get the number of frames for each embryo
n_frames_control = imagen_control_norm_minmax.shape[0]
n_frames_mutatnt1 = imagen_mutatnt1_norm_minmax.shape[0]
n_frames_mutatnt2 = imagen_mutatnt2_norm_minmax.shape[0]

# Create label vectors for each embryo type
labels_control = np.zeros(n_frames_control)      # label = 0 for control
labels_mutatnt1 = np.ones(n_frames_mutatnt1)       # label = 1 for mutant 1
labels_mutatnt2 = np.full(n_frames_mutatnt2, 2)    # label = 2 for mutant 2

# Concatenate the label vectors into a single label vector
y = np.concatenate((labels_control, labels_mutatnt1, labels_mutatnt2), axis=0)

print(f"Shape of label vector y: {y.shape}")
```

3. PCA (Principal Component Analysis)
Apply PCA on both normalized datasets:

Visualize the 2D PCA embedding colored by embryo type.
Compare the separation between classes for X_minmax and X_standard.
Plot the explained variance ratio and cumulative variance for each.
🧠 Tip: Use PCA(n_components=2) for plotting and PCA(n_components=50) to analyze cumulative variance.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Apply PCA with 2 components for plotting
pca_minmax = PCA(n_components=2)
X_minmax_pca = pca_minmax.fit_transform(X_minmax)

pca_standard = PCA(n_components=2)
X_standard_pca = pca_standard.fit_transform(X_standard)

# Create a figure with two subplots side-by-side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot the Min-Max normalized data
scatter1 = ax1.scatter(X_minmax_pca[:, 0], X_minmax_pca[:, 1], c=y, cmap='viridis', s=50, alpha=0.7)
ax1.set_title('PCA on Min-Max Normalized Data')
ax1.set_xlabel('Principal Component 1')
ax1.set_ylabel('Principal Component 2')
ax1.grid(True)
# Add legend
legend_labels = ['Control', 'Mutant 1', 'Mutant 2']
ax1.legend(handles=scatter1.legend_elements()[0], labels=legend_labels, title='Embryo Type')

# Plot the standardized data
scatter2 = ax2.scatter(X_standard_pca[:, 0], X_standard_pca[:, 1], c=y, cmap='viridis', s=50, alpha=0.7)
ax2.set_title('PCA on Standardized Data')
ax2.set_xlabel('Principal Component 1')
ax2.set_ylabel('Principal Component 2')
ax2.grid(True)
# Add legend
ax2.legend(handles=scatter2.legend_elements()[0], labels=legend_labels, title='Embryo Type')

plt.tight_layout()
plt.show()
```

```python
# Apply PCA with 50 components to analyze variance
pca_minmax_full = PCA(n_components=50)
pca_minmax_full.fit(X_minmax)

pca_standard_full = PCA(n_components=50)
pca_standard_full.fit(X_standard)

# Get the explained variance ratio and cumulative sum
explained_variance_minmax = pca_minmax_full.explained_variance_ratio_
cumulative_variance_minmax = np.cumsum(explained_variance_minmax)

explained_variance_standard = pca_standard_full.explained_variance_ratio_
cumulative_variance_standard = np.cumsum(explained_variance_standard)

# Plot the explained variance ratio and cumulative variance
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Min-Max Plot
ax1.bar(range(1, 51), explained_variance_minmax, alpha=0.5, label='Explained Variance Ratio')
ax1.step(range(1, 51), cumulative_variance_minmax, where='mid', label='Cumulative Variance')
ax1.set_xlabel('Principal Component')
ax1.set_ylabel('Explained Variance')
ax1.set_title('Explained Variance (Min-Max Normalized)')
ax1.legend()
ax1.grid(True)

# Standardization Plot
ax2.bar(range(1, 51), explained_variance_standard, alpha=0.5, label='Explained Variance Ratio')
ax2.step(range(1, 51), cumulative_variance_standard, where='mid', label='Cumulative Variance')
ax2.set_xlabel('Principal Component')
ax2.set_ylabel('Explained Variance')
ax2.set_title('Explained Variance (Standardized)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
